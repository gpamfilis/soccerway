```python
import json
import os
import pandas as pd
from soccerway import insert_date,insert_champ
```

```python
files = os.listdir('./json/')
```

```python
data = []
for f in files:
    try:
        with open("./json/"+f, "rb") as input_file:
            e = json.load(input_file)
        data.append([f[:-5],e])
    except Exception as e:
        continue
```

```python
dfs = []
for date, d in data:
    print('date')
    for k in  d.keys():
#         print d[k]
        fixed = insert_date(insert_champ(d[k],champ=k),date=date)
#         print(fixed)
        dfs.append(pd.DataFrame(fixed,columns=['date','championship','home_team',"score","away_team"]))
```

```python
final = pd.concat(dfs,ignore_index=True)

final.loc[:,"score"] = final.score.apply(lambda x: x.strip())

final.to_csv('./csv/data_soccerway_'+min(files)[:-5]+'_'+max(files)[:-5]+'.csv',index=False, encoding='utf8')
```
